### Connecting to Pinecone VectorDB

```python
pc = Pinecone(api_key="Your_Pinecone_API_Key")
```

```python
import requests
import numpy as np
from typing import List, Union

EURON_API_KEY = "Your_EURI_API_Key"

def generate_embeddings(texts: Union[str, List[str]]):
    # Always make the input a list
    if isinstance(texts, str):
        texts = [texts]

    url = "https://api.euron.one/api/v1/euri/embeddings"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {EURON_API_KEY}"
    }
    payload = {
        "input": texts,
        "model": "text-embedding-3-small"
    }

    response = requests.post(url, headers=headers, json=payload)
    data = response.json()

    # Convert each embedding to numpy array
    embeddings = [np.array(item["embedding"], dtype=np.float32) for item in data["data"]]

    # Return single vector OR batch of vectors
    return embeddings[0] if len(embeddings) == 1 else np.stack(embeddings)
```

```python
INDEX_NAME = "Your_Pinecone_table_name"
```

```python
pc.list_indexes()
```

### Creating Index (Table) in Pinecone VectorDB

```python
pc.create_index(
    name = INDEX_NAME,
    dimension = 1536,
    metric = "cosine",
    spec = ServerlessSpec(
        cloud = "aws",
        region = "us-east-1"
        )
    )
```

### Preparing data for embedding generation

```python
texts = [doc['text'] for doc in documents]
```

```python
texts
```

```python
doc_embeddings = generate_embeddings(texts)
```

```python
doc_embeddings
```

```python
documents, doc_embeddings
```

```python
zip(documents, doc_embeddings)
```

### Attaching metadata with embeddings

```python
vector_to_upsert = []

for doc, emb in zip(documents, doc_embeddings):
    
    metadata = {
        "category": doc["category"],
        "tag": doc["tag"],
        "difficulty": doc["difficulty"],
        "url": doc["url"],
        "text": doc["text"]
    }
    
    vector_item = {
        "id": doc["id"],
        "values": emb.tolist(),
        "metadata": metadata
    }
    
    vector_to_upsert.append(vector_item)
```

```python
vector_to_upsert[0]
```

```python
index = pc.Index(INDEX_NAME)
```

```python
index
```

### Inserting Embeddings to the index/table in Pinecone VectorDB

```python
index.upsert(vectors=vector_to_upsert)
```

### Testing

```python
query_text = "How to use Pinecone with Python?"

query_embedding = generate_embeddings(query_text)
```

```python
pi_response = index.query(
    vector=query_embedding.tolist(),
    top_k=3,
    include_metadata=True
)
```
